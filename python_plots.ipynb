```python
import matplotlib.pyplot as plt
import numpy as np

# Categories
categories = ['Average', 'General', 'Biomed', 'Clinics', 'STEM', 'Programming', 'Social media', 'Law', 'Finance', 'Transport']

# Data for each model (placeholders, replace with actual values from the image)
vicuna_7b = [14.2, 18.0, 9.8, 9.5, 8.9, 4.6, 14.6, 5.1, 24.9, 8.9]
vicuna_13b = [13.9, 21.5, 19.6, 13.2, 6.9, 5.7, 9.0, 12.1, 21.7, 11.5]
chatgpt = [34.9, 51.0, 38.1, 25.3, 18.6, 11.6, 31.3, 16.3, 52.8, 23.2]
uniner_7b = [41.7, 53.2, 51.5, 28.2, 29.3, 27.7, 41.6, 20.6, 58.7, 21.7]
uniner_13b = [43.4, 54.2, 51.2, 32.6, 32.7, 30.5, 44.4, 19.4, 60.9, 23.2]

# X locations of the groups
ind = np.arange(len(categories))  
width = 0.15

fig, ax = plt.subplots(figsize=(10, 8))

# Plotting data
rects1 = ax.bar(ind - width*2, vicuna_7b, width, label='Vicuna-7B')
rects2 = ax.bar(ind - width, vicuna_13b, width, label='Vicuna-13B')
rects3 = ax.bar(ind, chatgpt, width, label='ChatGPT')
rects4 = ax.bar(ind + width, uniner_7b, width, label='UniNER-7B')
rects5 = ax.bar(ind + width*2, uniner_13b, width, label='UniNER-13B')

# Adding text for labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('Domains')
ax.set_ylabel('F1 (%)')
ax.set_title('F1 Score by Model and Domain')
ax.set_xticks(ind)
ax.set_xticklabels(categories)
ax.legend()

# Adding the text with the percentage values
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Call the function to add labels
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
autolabel(rects5)

plt.tight_layout()

plt.show()
```

# Nested Pie Plots (Donut Plots)

```python
import matplotlib.pyplot as plt
import numpy as np

# Define data for the three layers
categories = ['Category A', 'Category B', 'Category C']
category_values = [900, 600, 1500]
category_colors = ['lightblue', 'lightgreen', 'lightcoral']

subcategories = ['A1', 'A2', 'B1', 'B2', 'C1', 'C2', 'C3']
subcategory_values = [300, 600, 300, 300, 500, 500, 500]
subcategory_colors = ['blue', 'dodgerblue', 'green', 'lime', 'red', 'darkorange', 'peachpuff']

items = ['A1a', 'A1b', 'A2a', 'A2b', 'B1a', 'B2a', 'B2b', 'C1a', 'C1b', 'C2a', 'C2b', 'C3a', 'C3b']
item_values = [150, 150, 300, 300, 300, 150, 150, 250, 250, 250, 250, 250, 250]
item_colors = ['aqua', 'deepskyblue', 'navy', 'royalblue', 'lightgreen', 'darkgreen', 'olive', 'salmon', 'orangered', 'chocolate', 'sienna', 'purple', 'violet']

fig, ax = plt.subplots()
r1 = 1.0
r2 = r1 + 1.0
r3 = r2 + 1.0
# Inner Ring: Categories
inner_ring, _ = ax.pie(category_values, radius=r1, colors=category_colors,
                       wedgeprops=dict(width=1, edgecolor='w'), startangle=90)
plt.setp(inner_ring, edgecolor='w')
for i, patch in enumerate(inner_ring):
    ang = (patch.theta2 - patch.theta1) / 2. + patch.theta1
    x = 0.5*r1 * np.cos(np.deg2rad(ang))
    y = 0.5*r1 * np.sin(np.deg2rad(ang))
    plt.text(x, y, categories[i], ha='center', va='center', fontsize=12)

# Middle Ring: Subcategories
middle_ring, _ = ax.pie(subcategory_values, radius=r2, colors=subcategory_colors,
                        wedgeprops=dict(width=1, edgecolor='w'), startangle=90)
plt.setp(middle_ring, edgecolor='w')
for i, patch in enumerate(middle_ring):
    ang = (patch.theta2 - patch.theta1) / 2. + patch.theta1
    x = (r1 + 0.5*(r2 - r1)) * np.cos(np.deg2rad(ang))
    y = (r1 + 0.5*(r2 - r1)) * np.sin(np.deg2rad(ang))
    plt.text(x, y, subcategories[i], ha='center', va='center', fontsize=12)

# Outer Ring: Items
outer_ring, _ = ax.pie(item_values, radius=r3, colors=item_colors,
                       wedgeprops=dict(width=1, edgecolor='w'), startangle=90)
plt.setp(outer_ring, edgecolor='w')
for i, patch in enumerate(outer_ring):
    ang = (patch.theta2 - patch.theta1) / 2. + patch.theta1
    x = (r2 + 0.5*(r3 - r2)) * np.cos(np.deg2rad(ang))
    y = (r2 + 0.5*(r3 - r2)) * np.sin(np.deg2rad(ang))
    plt.text(x, y, items[i], ha='center', va='center', fontsize=12)

plt.show()
```
